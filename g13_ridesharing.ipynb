```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("data/g13_ridesharing.xlsx")
conn = sqlite3.connect("data/g13_ridesharing.db")
```

```python
cars = df.loc[:, ['id_car', 'description', 'id_company', 'car_type']].copy()
cars = cars.drop_duplicates(subset='id_car', keep='first')
cars['id_car'] = cars['id_car'].astype(int)
cars['id_company'] = cars['id_company'].astype(int)
cars['car_type'] = cars['car_type'].astype(int)
cars = cars.rename(columns={'id_car': 'id'})
```

```python
cars.to_sql("Car", conn, if_exists="append", index=False)
```

```python
pd.read_sql("SELECT * FROM Car LIMIT 20", conn)
```

```python
companies = df.loc[:, ['id_company', 'name', 'begin_date']].copy()
companies = companies.drop_duplicates(subset='id_company', keep='first')
companies['id_company'] = companies['id_company'].astype(int)
companies = companies.rename(columns={'id_company': 'id'})
```

```python
companies.to_sql("Company", conn, if_exists="append", index=False)
```

```python
pd.read_sql("SELECT * FROM Company LIMIT 20", conn)
```

```python
contracts = df.loc[:, ['id_contract', 'contract_start', 'contract_end', 'id_company', 'id_driver']].copy()
contracts = contracts.drop_duplicates(subset='id_company', keep='first')
contracts['id_contract'] = contracts['id_contract'].astype(int)
contracts['id_company'] = contracts['id_company'].astype(int)
contracts['id_driver'] = contracts['id_driver'].astype(int)
contracts = contracts.rename(columns={'id_contract': 'id'})
```

```python
contracts.to_sql("Contract", conn, if_exists="append", index=False)
```

```python
pd.read_sql("SELECT * FROM Contract LIMIT 20", conn)
```

```python
customers = df.loc[:, ['id_customer', 'name_customer', 'email', 'phone', 'date_of_birth']].copy()
customers = customers.drop_duplicates(subset='id_customer', keep='first')
customers['id_customer'] = customers['id_customer'].astype(int)
customers['phone'] = customers['phone'].astype(int)
customers = customers.rename(columns={'id_customer': 'id'})
customers = customers.rename(columns={'name_customer': 'name'})
```

```python
customers.to_sql("Customer", conn, if_exists="append", index=False)
```

```python
pd.read_sql("SELECT * FROM Customer LIMIT 20", conn)
```

```python
drivers = df.loc[:, ['id_driver', 'nickname', 'driver_type']].copy()
drivers = drivers.drop_duplicates(subset='id_driver', keep='first')
drivers['id_driver'] = drivers['id_driver'].astype(int)
drivers = drivers.rename(columns={'id_driver': 'id'})
```

```python
drivers.to_sql("Driver", conn, if_exists="append", index=False)
```

```python
pd.read_sql("SELECT * FROM Driver LIMIT 20", conn)
```

```python
rides = df.loc[:, ['id_ride', 'id_company', 'id_driver', 'id_customer', 'id_car', 'origin', 'destination', 'ride_date', 'amount' ]].copy()
rides = rides.drop_duplicates(subset='id_ride', keep='first')
rides['id_ride'] = rides['id_ride'].astype(int)
rides['id_company'] = rides['id_company'].astype(int)
rides['id_driver'] = rides['id_driver'].astype(int)
rides['id_customer'] = rides['id_customer'].astype(int)
rides['id_car'] = rides['id_car'].astype(int)
rides['amount'] = rides['amount'].astype(int)
rides = rides.rename(columns={'id_ride': 'id'})
```

```python
rides.to_sql("Ride", conn, if_exists="append", index=False)
```

```python
pd.read_sql("SELECT * FROM Ride LIMIT 20", conn)
```
